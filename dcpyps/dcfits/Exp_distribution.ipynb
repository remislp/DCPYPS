```python
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x = np.logspace(np.log10(0.0001), np.log10(10), 1024)
y = np.exp(-x)
```

```python
fig, ax  = plt.subplots(1,1, figsize=(5,3))
ax.plot(x, y)
ax.set_xlabel('X')
ax.set_ylabel('Y')
#ax.set_xscale('log')
```

```python
fig, ax  = plt.subplots(1,1, figsize=(5,3))
ax.plot(x, y)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_xscale("log", nonposx='clip')
ax.set_yscale("log", nonposy='clip')
```

```python
fig, ax  = plt.subplots(1,1, figsize=(5,3))
ax.plot(x, np.sqrt(y*x))
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_xscale('log')
```
